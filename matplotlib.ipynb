```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = np.linspace(0, 5, 11)
```

```python
x
```

```python
y = x**2
```

```python
y
```

```python
#plt.plot(x, y, color='r')
# ou 
plt.plot(x, y, 'b--')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Título do Gráfico')
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r--')
plt.subplot(1, 2, 2)
plt.plot(y, x, 'g*-')
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x, y)
axes.set_xlabel('Eixo X')
axes.set_title('Título')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.3, 0.3])

axes1.plot(x, y)
axes1.set_xlabel('Eixo X')
axes1.set_title('Título')

axes2.plot(y, x)
```

```python
fig, ax = plt.subplots()
ax.plot(x, x**3, 'b')
ax.plot(x, x**4, 'g-.')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Título')
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
```

```python
for axe in ax:
    axe.plot(x, y, 'b')
    axe.set_title('Título')
fig
```
