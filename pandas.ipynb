```python
# carregando os dados
# path de localização do arquivo
DATA_PATH = 'dados/kc_house_data.csv'

# dataframe = df

df = pd.read_csv(DATA_PATH)
```

```python
# Realizando consultas no dataframe método value_counts()

print('Quartos/Imóveis')
print(pd.value_counts(df['bedrooms']))
```

```python
# Consultar Imóveis com 3 quartos método loc[]

df.loc[df['bedrooms'] == 3].head()
```

```python
# consultar imóveis com 3 quartos e com mais de 2 banheiros  condições - método loc[() & ()]

df.loc[(df['bedrooms'] == 3) & (df['bathrooms'] > 2)].head() 
```

```python
# método sort_values() - ordena valores por coluna

df.sort_values(by='price').head()
```

```python
# método count() conta o número de linhas

df[df['bedrooms'] == 4].count()
```

## Alterando o Dataframe

```python
df.head()
```

```python
# adicionando uma coluna

df['size'] = (df['bedrooms'] * 20)
```

```python
df.head()
```

```python
# criando uma função para processamento de dados

def categoriza(s):
    if s >= 80:
        return 'Big'
    elif s >= 60:
        return 'Medium'
    elif s >= 40:
        return 'Small'
```

```python
# criando uma nova coluna baseada no resultado obtido pelo processamento de dados

df['cat_size'] = df['size'].apply(categoriza)
df.head()
```

```python
df['cat_size'].head()
```

```python
# ver a distribuição dos valores na coluna cat_size

pd.value_counts(df['cat_size'])
```

```python
# o método drop é utilizado para excluir dados no dataframe
# o parâmetro axis = 1 define a exclução de uma coluna e axis = 0, de uma linha
# o parâmetro inplace define que a alteração irá modificar o objeto em memória

df.drop(['cat_size'], axis = 1, inplace = True)
```

```python
df.head()
```

```python
# excluindo uma coluna

df.drop(['size'], axis=1, inplace=True)
```

```python
df.head()
```

```python
# excluíndo linhas conforme uma condição lógica

# excluindo linhas onde bedrooms for igual a 0

df.drop(df[df['bedrooms'] == 0].index, inplace=True)
```

```python
# excluindo linhas nas quais o número de quarto seja maior que 30

df.drop(df[df['bedrooms'] > 30].index, inplace=True)
```

# Missing Values
Valores faltantes

```python
# carregando os dados
# path de localização do arquivo
DATA_PATH = 'dados/kc_house_data.csv'

# dataframe = df

df = pd.read_csv(DATA_PATH)
```

```python
# é possível fazer a verificação de valores faltantes através da método isnull()

df.isnull().sum()
```

```python
# removendo todas as linhas que tem pelo menos um valor faltante
df.dropna(inplace=True)
```

```python
# removendo valores faltantes quando os mesmos estão presentes em todas as colunas
df.dropna(how='all', inplace=True)
```

```python
# preechendo os valores faltantes com a média de outros valores
df['floors'].fillna(df['floors'].mean(), inplace=True)
```

```python
# preenchendo os valores faltantes com um valor específico
df['bedrooms'].fillna(1, inplace=True)
```

```python
df.isnull().sum()
```
