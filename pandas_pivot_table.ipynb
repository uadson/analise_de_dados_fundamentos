## Pivot Table

    * São úteis para agrupamento de dados
    
    Parâmetros:
    
        index : variáveis(colunas) que serão agrupadas
        aggfunc : função de agregação, pode ser um count ou mean por exemplo
        margins : se True, calcula a quantidade total no final da tabela

```python
import pandas as pd
```

```python
data = pd.read_csv(r'dados/kc_house_data.csv')
```

```python
data.pivot_table('id', index=['waterfront', 'floors'], aggfunc = 'count', margins = True)
```

### Crosstab

    * Avaliação cruzada dos dados

```python
pd.crosstab(data['bedrooms'], data['condition'])
```

```python
# plotando a tabulação cruzada

%matplotlib notebook
table = pd.crosstab(data['bedrooms'], data['condition'])

table.plot(kind='bar', width=1.0, color=['red', 'yellow', 'orange', 'blue', 'green'], title='Condições por Quartos', grid=False)
```
