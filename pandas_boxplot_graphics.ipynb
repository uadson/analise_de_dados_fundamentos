### Gráficos do tipo BoxPlot

```python
import pandas as pd
```

```python
df = pd.read_csv(r'dados/kc_house_data.csv')
```

```python
df.head()
```

```python
import matplotlib

%matplotlib notebook
matplotlib.style.use('ggplot')
df.boxplot(column='bedrooms')
```

```python
# Estilos disponíveis

matplotlib.style.available
```

```python
%matplotlib notebook
df.boxplot(column='price', by='bedrooms')
```

### Correlação de Variáveis

```python
df.corr()
```

### Correlação Positiva e Negativa entre as variáveis

```python
df.corr('spearman')
```

```python
# Verificando a correlação somente entre algumas colunas

colunas = 'bedrooms bathrooms sqft_living floors waterfront grade price'.split()

df[colunas].corr()
```

```python
# plotando as correlações

%matplotlib notebook

colunas = 'sqft_living waterfront grade price'.split()

df[colunas].corr().plot()
```
